# <font color='blue'>3. ffmpeg: Scale pictures </font>


adjust the scale = __512x512__

Important replace __*XXX_path_XXX*__ with your directory


```python
%%bash
sh
for file in *; do
      ffmpeg -i \
      \
      /Users/arminarndt/ffmpeg_workstation/1_raw_images/"$file" \
      \
      -vf scale=512:512 \
      \
      /Users/arminarndt/ffmpeg_workstation/2_edited_images/"${file%}"
done 
```

# <font color='blue'>4. ffmpeg: JPEG --> PNG</font>

# <font color='blue'>Bulk re-name files</font>

change directory to the folder containing the images

                          

```python
cd .././ffmpeg_workstation/1_raw_images
```

Option 1
-----------------------   
If you have 1 folders with JPGs and you want to rename them

                      ----> 
        folder                   folder
            *.jpg                     00001.jpg
            *.jpg                     00002.jpg
            *.jpg                     00003.jpg
            *.jpg                     00004.jpg
            *.jpg                     00005.jpg
            *.jpg                     00006.jpg
            *.jpg                     00007.jpg
            *.jpg                     00008.jpg
            *.jpg                        ...jpg


```python
%%bash
sh
a=1
for i in *jpg; do
  new=$(printf "%05d.jpg" "$a")
  mv -i -- "$i" "$new"
  let a=a+1
done
```

 
Option 2 
-----------------------   
If you have many folders and you want to collect all the JPGs from the sub-directory and rename them

                      ----> 
    folder                       folder
        folder                      00001.jpg
            *.jpg                     00002.jpg
            *.jpg                     00003.jpg
            *.jpg                     00004.jpg
            *.jpg                     00005.jpg
        folder                      00006.jpg
            *.jpg                     00007.jpg
            *.jpg                     00008.jpg
            *.jpg                        ...jpg
            *.jpg                     folder (now empty)
                                      folder (now empty)

```python
%%bash
sh
a=1
for i in */*jpg; do
  new=$(printf "%05d.jpg" "$a")
  mv -i -- "$i" "$new"
  let a=a+1
done
```
